```python
import pandas as pd
```

```python
df = pd.read_csv("Datasets/tb.csv")
df.dropna(inplace=True)
df['new_sp_m014'] = df.iloc[:,3:6].sum(axis=1)
df = df.reset_index(drop=True)
df = df.drop(columns=['new_sp','new_sp_m04','new_sp_m514','new_sp_mu','new_sp_fu'])
df.head(20)
```

```python
df = pd.melt(df, id_vars=['iso2', 'year'], value_vars=list(df.columns)[2:],
             var_name='column', value_name='cases')
df = df.dropna()

df['sex'] = df['column'].str[7]
df['age'] = df['column'].str[8:].map({
    '014': '0-14',
    '1524': '15-24',
    '2534': '25-34',
    '3544': '35-44',
    '4554': '45-54',
    '5564': '55-64',
    '65': '65+'
})
df = df[['iso2', 'year', 'cases', 'sex', 'age']]
df.sort_values(by=['iso2','year'])
df.to_csv('Datasets/tb-tidy.csv', index=False)
```
