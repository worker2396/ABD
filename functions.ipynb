Zadanie 3

```python
def foo(x):
    return x**2 + 5
x1 = np.linspace(-1,1,50)
x2 = np.linspace(-6,6,100)
x3 = np.linspace(0,5,100)
```

Stworzono definicję funkcji wraz z przedziałami jej argumentów.

```python
plt.plot(x1, foo(x1))
plt.title("X1")
```

```python
plt.plot(x2, foo(x2))
plt.title("X2")
```

```python
plt.plot(x3, foo(x3))
plt.title("X3")
```

Zostały sporządzone wykresy funkcji z poszczególnymi ciągami argumentów x za pomocą funkcji plot z biblioteki matplotlib.
